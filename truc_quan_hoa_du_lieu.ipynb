Trung bình di động (Moving Average) là một kỹ thuật thống kê được sử dụng để làm mịn dữ liệu thời gian bằng cách tính trung bình của các giá trị trong một khoảng thời gian liên tiếp. Điều này giúp giảm nhiễu và làm nổi bật xu hướng chung trong dữ liệu.

Công thức: SUM = tổng giá qua x ngày; MA = SUM / x

```python
MA_days = [10, 20, 30]

for ma in MA_days:
    ma_str = "MA: {}".format(ma)
    AAPL_2023[ma_str] = AAPL_2023['close'].rolling(ma).mean()
    AMZN_2023[ma_str] = AMZN_2023['close'].rolling(ma).mean()
    TSLA_2023[ma_str] = TSLA_2023['close'].rolling(ma).mean()

AAPL_2023.head()
```

```python
data_list = [AAPL_2023, AMZN_2023, TSLA_2023]
company_list = ['AAPL_2023', 'AMZN_2023', 'TSLA_2023']

for i in range(len(data_list)):   
    plot_graphic(data_list[i], company_list[i])
```

Nếu đang xem xét việc đầu tư dài hạn, thì việc sử dụng đường trung bình di động sẽ tốt hơn. Lý do là vì đường trung bình di động mịn hơn sẽ ít bị ảnh hưởng bởi những biến động giá đột ngột.

```python
# Tính toán phần giá trị phần trăm trả về và vẽ biểu đồ phân phối

for i in range(len(data_list)):
    data_list[i]['Daily Returns'] = data_list[i]['close'].pct_change()
    sns.displot(data_list[i]['Daily Returns'].dropna(), bins=50, color='blue', kde=True)
    plt.title(company_list[i])
    plt.savefig(str(company_list[i]) + "_returns" + ".png") 
    plt.show()
```

Sự tương quan giữa các giá trị phần trăm trả về của cổ phiếu

```python
stock_returns = pd.DataFrame(data=np.array([data['Daily Returns'] for data in data_list]).T, columns=company_list)
stock_returns.head()
```

```python
sns.pairplot(stock_returns.dropna())
```

```python
# Xây dựng ma trận tương quan
corr = stock_returns.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
plt.figure(figsize=(10, 10))
sns.heatmap(corr, mask=mask,  square=True, linewidths=.5, annot=True)
plt.savefig("Correlation_heatmap.png") 
plt.show()
```

```python
def draw_jointplot(data):
    grid = sns.PairGrid(data.dropna())
    grid.map_diag(sns.histplot, bins=40, kde=True)
    grid.map_lower(sns.regplot)
    grid.map_upper(sns.kdeplot)
    plt.savefig("jointplot.png")
```

```python
draw_jointplot(stock_returns)
```

Từ những biểu đồ trên ta thấy có sự tương quan giữa các cổ phiếu, nhưng nó chỉ nằm trong khoảng [0.35, 0.45]. Với cặp AAPL/TSLA và cặp AAPL/AMZN có độ tương quan giá trị Phần trăm trả về cao nhất. Cặp AMZN/TSLA có độ tương quan giá trị Phần trăm trả về thấp nhất.
